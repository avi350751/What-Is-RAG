```python
from langchain_community.document_loaders.pdf import PyPDFLoader
from pathlib import Path
from pprint import pprint
```

```python
# Create the path for pdf file
pdf_path = Path("../1_document_loaders/Attention_Is_All_You_Need.pdf")
pdf_path.exists() # Check if the file exists
```

```python
file_path=pdf_path.as_posix() #get the path in string format
```

```python
##create a pdf loader
loader = PyPDFLoader(file_path=file_path, mode='page')
```

```python
# load the documents
documents = loader.load()
```

```python
# number of documents
len(documents)
```

```python
# fetch the first document
print(documents[0].page_content)
```

```python
# metadat
pprint(documents[0].metadata)
```

```python
from langchain_community.document_loaders.parsers import TesseractBlobParser
```

```python
# Create pypdf instance which can extract images

pypdf_image_loader = PyPDFLoader(file_path=file_path, mode='page',
                                 extract_images=True,
                                 images_parser=TesseractBlobParser(),
                                 images_inner_format='html-img')
```

```python
# load documents with images
documents_with_images = pypdf_image_loader.load()
```

```python
len(documents_with_images)
```

```python
print(documents_with_images[2].page_content)
```
