```python
from langchain_openai import OpenAIEmbeddings
from langchain_community.document_loaders.pdf import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from pathlib import Path
from dotenv import load_dotenv

load_dotenv()
```

```python
# small model
# length of the embedding vector is 1536

embeddings = OpenAIEmbeddings(model="text-embedding-3-small")

#Create emeddings for a generic query
query = "What is the capital of Estonia?"

#embed query
query_vector = embeddings.embed_query(query)

print(f"Query is: {query}")
print(f"Length of the query vector: {len(query_vector)}")
print(f"Query vector: {query_vector[:10]}")
```

```python
# large model
# length of the embedding vector is 3072

embeddings_large = OpenAIEmbeddings(model="text-embedding-3-large")

#Create emeddings for a generic query
query = "In which country is the current FIFA world cup happening?"

#embed query
query_vector = embeddings_large.embed_query(query)

print(f"Query is: {query}")
print(f"Length of the query vector: {len(query_vector)}")
print(f"Query vector: {query_vector[:10]}")
```

```python
## Embed multiple documents
documents = [
    "The capital of Estonia is Tallinn.",
    "The capital of France is Paris.",
    "The capital of Germany is Berlin."
]

#embed documents
document_vectors = embeddings.embed_documents(documents)

for i, (doc, vec) in enumerate(zip(documents, document_vectors)):
    print(f"Document {i + 1}: {doc}")
    print(f"Length of the document vector: {len(vec)}")
    print(f"Document vector: {vec[:10]}")
    print()
```

### Let's embed a pdf document

```python
pdf_path = Path("../1_document_loaders/Attention_Is_All_You_Need.pdf")
loader = PyPDFLoader(file_path=pdf_path.as_posix(),mode='page')
documents = loader.load()

len(documents)
```

```python
#split documents
# 
chunks = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
).split_documents(documents)

len(chunks)
```

```python
# embed documents
text_doc = [doc.page_content for doc in chunks]

print(len(text_doc))

docu_embeddings = embeddings.embed_documents(text_doc)

print(len(docu_embeddings))
```
